## Predictive Modeling: Estimating Tourism Revenue with Linear Regression

A linear regression model is built to predict `tourism_revenue_inr` using tourism-related features such as spending behavior, infrastructure, connectivity, sentiment, and safety.

Steps:
- Features are standardized to prevent scale dominance.
- Data is split into training and testing sets (80/20).
- Model performance is evaluated using R² and Mean Squared Error (MSE).
- Coefficients are analyzed to identify impactful predictors.
- Residuals are visualized to assess fit quality and linearity assumptions.

This helps determine whether a linear approach is suitable and which features influence revenue the most.


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

# Define features and target
features = [
    'num_tourists', 'avg_spending_per_visitor', 'gdp_contribution_percent',
    'num_hotels', 'hotel_occupancy_rate', 'avg_room_tariff', 'num_airports',
    'num_tourist_info_centers', 'avg_attraction_rating', 'num_cultural_events',
    'connectivity_score', 'avg_travel_cost_to_state', 'num_trains_per_day',
    'road_quality_index', 'avg_sentiment_score', 'total_review_count',
    'safety_rating'
]
X = df[features]
y = df['tourism_revenue_inr']

# Scaling features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=1)

# Train model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse:.2f}")
print(f"R-squared Score: {r2:.2f}")

# Feature importance (coefficients)

coef_df['Effect'] = ['Positive' if c >= 0 else 'Negative' for c in coef_df['Coefficient']]

# Barplot with hue assigned and legend hidden
plt.figure(figsize=(10, 5))
sns.barplot(
    data=coef_df,
    x='Coefficient',
    y='Feature',
    hue='Effect',
    palette='coolwarm',
    dodge=False,
    legend=False
)
plt.title("Feature Importance (Linear Coefficients)")
plt.tight_layout()
plt.show()


# Residual plot
residuals = y_test - y_pred
plt.figure(figsize=(6, 4))
sns.scatterplot(x=y_pred, y=residuals, color='darkred', edgecolor='white')
plt.axhline(0, linestyle='--', color='gray')
plt.title("Residuals vs Predicted Revenue")
plt.xlabel("Predicted Revenue")
plt.ylabel("Residuals")
plt.tight_layout()
plt.show()

```

```python
# Plot Actual vs Predicted
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_test, y=y_pred, color='purple')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel('Actual Revenue')
plt.ylabel('Predicted Revenue')
plt.title('Actual vs Predicted Tourism Revenue')
plt.grid(True)
plt.tight_layout()
plt.show()
```
